```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for col in df.columns:
  if df[col].dtype!='object':
    sns.boxplot(df[col])
    plt.show()
```

```python
list=['age','fnlwgt','education-num']
```

```python
for col in list:
  q1=df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  iqr=q3-q1
  UL=q3+1.5*iqr
  LL=q1-1.5*iqr
  df=df[(df[col]<=UL)&(df[col]>=LL)]
```

```python
# Rename your list variable to something else, like 'columns_to_check'
columns_to_check = ['age','fnlwgt','education-num']

# Now, use this new variable in your for loop
for col in columns_to_check:
  q1=df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  iqr=q3-q1
  UL=q3+1.5*iqr
  LL=q1-1.5*iqr
  df=df[(df[col]<=UL)&(df[col]>=LL)]
```

```python
df['annual_income']=df['annual_income'].map({'<=50K':0,'>50K':1})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for col in df.columns:
  if df[col].dtype=='object':
    df[col]=le.fit_transform(df[col])
    print(le.classes_)
```

```python
x=df.drop('annual_income',axis=1)
y=df['annual_income']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
#
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
